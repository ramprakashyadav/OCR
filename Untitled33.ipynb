```python
import matplotlib.pyplot as plt
import cv2
import easyocr
from pylab import rcParams
from IPython.display import Image
rcParams['figure.figsize'] = 210, 297
```

```python
reader = easyocr.Reader(['en'])
```

```python
Image("Image.03.jpg")
```

```python
output = reader.readtext('Image.03.jpg')
```

```python
output
```

```python
cord = output[10][0]
cord1 = output[8][0]
cord2 = output[-3][0]
cord3 = output[-5][0]
cord4 = output[-6][0]
```

```python
x_min, y_min = [int(min(idx)) for idx in zip(*cord)]
x_min1, y_min1 = [int(min(idx)) for idx in zip(*cord1)]
x_min2, y_min2 = [int(min(idx)) for idx in zip(*cord2)]
x_min3, y_min3 = [int(min(idx)) for idx in zip(*cord3)]
x_min4, y_min4 = [int(min(idx)) for idx in zip(*cord4)]
```

```python
x_max, y_max = [int(max(idx)) for idx in zip(*cord)]
x_max1, y_max1 = [int(max(idx)) for idx in zip(*cord1)]
x_max2, y_max2 = [int(max(idx)) for idx in zip(*cord2)]
x_max3, y_max3 = [int(max(idx)) for idx in zip(*cord3)]
x_max4, y_max4 = [int(max(idx)) for idx in zip(*cord4)]
```

```python
image = cv2.imread('Image.03.jpg')
cv2.rectangle(image,(x_min,y_min),(x_max,y_max),(0,0,255),2)
cv2.rectangle(image,(x_min1,y_min1),(x_max1,y_max1),(0,0,255),2)
cv2.rectangle(image,(x_min2,y_min2),(x_max2,y_max2),(0,0,255),2)
cv2.rectangle(image,(x_min3,y_min3),(x_max3,y_max3),(0,0,255),2)
cv2.rectangle(image,(x_min4,y_min4),(x_max4,y_max4),(0,0,255),2)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
```
